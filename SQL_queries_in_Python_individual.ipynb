### 1 Select the city that has the most average review count of its restaurant

We are passionate food enthusiasts with a dream of creating a unique dining experience, embarks on a data-driven journey to uncover the ideal restaurant concept.

With our first SQL query, we discover that **North Redington Beac**h in **Florida**, a picturesque coastal city, has the highest average review count for its restaurants.

```python
from sqlalchemy import create_engine, Column, Integer, String, Float, func
from sqlalchemy.orm import sessionmaker
from sqlalchemy.ext.declarative import declarative_base

# Define the database connection URL
db_url = 'mysql://B03:B03password@testproject.mysql.database.azure.com/B03'

# Create a SQLAlchemy engine to connect to the database
engine = create_engine(db_url)

# Create a session
Session = sessionmaker(bind=engine)
session = Session()

# Define a SQLAlchemy model for your "business" table
Base = declarative_base()

class Business(Base):
    __tablename__ = 'business'
    business_id = Column(String, primary_key=True)
    name = Column(String)
    city = Column(String)
    state = Column(String)
    review_count = Column(Integer)
    is_open = Column(Integer)
    attributes = Column(String)

# Fetch data from the "business" table
query = session.query(
    Business.business_id,
    Business.city,
    Business.state,
    func.round(func.avg(Business.review_count)).label('average_review_count')
) \
.filter(Business.review_count > 100) \
.group_by(Business.business_id, Business.city, Business.state) \
.order_by(func.avg(Business.review_count).desc()) \
.limit(50)

results = query.all()

# Print the results
for row in results:
    print(row)
```

###2 Examine all the businesses in that city to choose the category to open

###3 Get the average star of all the reviews for the selected resaturant

```python
# Define a SQLAlchemy model for your "review" table
class Review(Base):
    __tablename__ = 'review'
    id = Column(Integer, primary_key=True)
    business_id = Column(String)
    user_id = Column(String)
    stars = Column(Float)
    useful = Column(Integer)
    text = Column(String)
    date = Column(String)
    # Define other columns as needed


    # Subquery to get the business_id for the specified city
    subquery = session.query(Business.business_id) \
        .filter(Business.city == "North Redington Bch") \
        .order_by(Business.review_count.desc()) \
        .limit(1) \
        .subquery()

    # Main query to calculate the average of "stars" and count rows in the "review" table
    query = session.query(
        func.avg(Review.stars).label('average_stars'),
        func.count().label('row_count')
    ) \
    .filter(Review.business_id == subquery)

    result = query.first()
    print("")
    print("Average Stars:", result.average_stars)
    print("Row Count:", result.row_count)
```

###6 order the users who are influential based on the positive reviews from the selected restaurant

```python
# Create a declarative base
Base = declarative_base()

class User(Base):

    __tablename__ = 'user'

    user_id = Column(String, primary_key=True)
    name = Column(String)
    fans = Column(Integer)
    review_count = Column(Integer)
    useful = Column(Integer)
    average_stars = Column(Double)
    # Define other columns as needed

    # Subquery to get the business_id for the specified city with the most review_count
    subquery_business = session.query(Business.business_id) \
        .filter(Business.city == "North Redington Bch") \
        .order_by(Business.review_count.desc()) \
        .limit(1) \
        .subquery()

    # Subquery to calculate the average of "stars" for the specified business
    avg_stars_subquery = session.query(func.avg(Review.stars)) \
        .filter(Review.business_id == subquery_business) \
        .subquery()

    # Subquery to get the user_id for reviews that meet specific criteria
    subquery_user = session.query(Review.user_id) \
        .filter(Review.business_id == subquery_business) \
        .filter(Review.useful > 0) \
        .filter(Review.stars > avg_stars_subquery) \
        .order_by(Review.stars.desc()) \
        .subquery()

    # Main query to fetch user information for users in the subquery_user results
    query = session.query(User.user_id, User.name, User.fans, User.review_count, User.useful, User.average_stars, Review.business_id) \
        .join(Review, User.user_id == Review.user_id) \
        .filter(User.user_id.in_(subquery_user)) \
        .filter(Review.business_id == subquery_business) \
        .order_by(User.fans.desc())

    results = query.all()

    for row in results:
        print("User ID:", row.user_id)
        print("Name:", row.name)
        print("Fans:", row.fans)
        print("Review Count:", row.review_count)
        print("Useful:", row.useful)
        print("Average Stars:", row.average_stars)
        print("Business ID:", row.business_id)

```
