##3.1 Select the city that has the most average review count of its restaurant


We are passionate food enthusiasts embarking on a data-driven journey to draft our brilliant restaurant plan.

First of all, we want to select an ideal location that most people would go for restaurants.

With our first SQL query, we discover that **North Redington Beach in Florida**, a picturesque coastal city, has the highest average review count for its restaurants.

```python
%%sql
SELECT city, state, ROUND(AVG(review_count),0) AS average_review_count
FROM business
WHERE review_count > 100
GROUP BY city, state
ORDER BY average_review_count DESC
limit 50
```

##3.3 Get the average star of all the reviews for the selected resaturant



Continuing our journey, we calculate the **average star rating** of all the reviews for the selected restaurant category, BBQ Grill, in North Redington Beach.

With each star awarded by diners, we gain a clearer understanding of the standards we must meet to create an unforgettable dining experience.

```python
%%sql
select avg(stars), count(*)
from review
where business_id =
  (SELECT business_id
  FROM business
  WHERE city="North Redington Bch"
  ORDER BY review_count desc
  limit 1)
```

##3.4 Get the useful user id who rates over the average for the restuarant and their review contents



In the pursuit of culinary excellence, our fourth query uncovers these influential diners' insightful review contents. Their opinions will help us keep valuable features to shape our own restaurant.

```python
%%sql
SELECT user_id, text, stars
FROM review
WHERE business_id =
          (SELECT business_id
           FROM business
           WHERE city="North Redington Bch"
           ORDER BY review_count desc
           limit 1)
      AND useful > 0
      AND stars >
          (SELECT AVG(stars)
           FROM review
           WHERE business_id =
                      (SELECT business_id
                       FROM business
                       WHERE city="North Redington Bch"
                       ORDER BY review_count desc
                       limit 1)
           AND useful > 0
          )
ORDER BY stars DESC
LIMIT 50;
```

##3.5 Get the useful user id who rates 0 or 1 star for the restuarant and their review contents



Not shying away from constructive criticism, our fifth query identifies users who rate the selected restaurant with 0 or 1 star.

Their reviews, though unfavorable, provide us with a wealth of feedback. We see these reviews as an opportunity to understand and address potential shortcomings.

```python
%%sql
SELECT user_id, text, stars
FROM review
WHERE business_id =
          (SELECT business_id
           FROM business
           WHERE city="North Redington Bch"
           ORDER BY review_count desc
           limit 1)
    AND useful > 0
    AND stars =1 or stars=0
ORDER BY stars DESC
LIMIT 50;
```

##3.6 order the users who are influential based on the positive reviews from the selected restaurant



We believe in giving credit where it's due. Our sixth query allows us to recognize the most influential users based on their positive reviews of the selected restaurant. These users become our champions, and we looks forward to having them enjoy our future establishment.

```python
%%sql
SELECT u.user_id, u.name, u.fans, u.review_count, u.useful, u.average_stars, r.business_id
FROM user AS u
JOIN review AS r ON u.user_id = r.user_id
WHERE u.user_id IN (
    SELECT r.user_id
    FROM review AS r
    WHERE r.business_id = (
        SELECT business_id
        FROM business
        WHERE city = "North Redington Bch"
        ORDER BY review_count DESC
        LIMIT 1
    )
    AND r.useful > 0
    AND r.stars > (
        SELECT AVG(stars)
        FROM review
        WHERE business_id = (
            SELECT business_id
            FROM business
            WHERE city = "North Redington Bch"
            ORDER BY review_count DESC
            LIMIT 1
        )
    )
    ORDER BY r.stars DESC
)
AND r.business_id = (
    SELECT business_id
    FROM business
    WHERE city = "North Redington Bch"
    ORDER BY review_count DESC
    LIMIT 1
)
ORDER BY u.fans DESC;
```

##3.7 order the users who are influential based on the negative reviews from the selected restaurant



No stranger to taking criticism as a path to growth, our seventh query helps us identify influential users based on their negative reviews. We views this as an opportunity to learn from past restaurant shortcomings and ensure our future establishment exceeds expectations.

```python
%%sql
SELECT u.user_id, u.name, u.fans, u.review_count, u.useful, u.average_stars, r.business_id
FROM user AS u
JOIN review AS r ON u.user_id = r.user_id
WHERE u.user_id IN (
    SELECT user_id
    FROM review
    WHERE business_id = (
        SELECT business_id
        FROM business
        WHERE city = "North Redington Bch"
        ORDER BY review_count DESC
        LIMIT 1
    )
    AND useful > 0
    AND (stars = 1 OR stars = 0)
)
AND r.business_id = (
    SELECT business_id
    FROM business
    WHERE city = "North Redington Bch"
    ORDER BY review_count DESC
    LIMIT 1
)
ORDER BY u.fans DESC;

```

##3.8 Get the average of stars in the latest year of the restaurant

To stay on top of changing tastes and expectations, our eighth query computes the average star rating of the restaurant in the latest year. This data provides us with a crucial insight into how the restaurant has been performing more recently, helping us adapt to evolving culinary trends.

```python
%%sql
SELECT business_id, AVG(stars) AS avg_stars_recent
FROM review
WHERE DATE BETWEEN '2017-01-01' AND '2018-12-31'
AND business_id = (
    SELECT business_id
    FROM business
    WHERE city = 'North Redington Bch'
    ORDER BY review_count DESC
    LIMIT 1
)
GROUP BY business_id;
```

##3.9 Check whether the selected restauant is still operating



The ninth query takes us on an exploration of the restaurant's current status. We confirm whether our selected restaurant is still operating or if it has closed down. This knowledge is vital for us to understand the challenges that might have led to any closures in the future.

```python
%%sql
SELECT name, is_open
from business
WHERE business_id =
 (SELECT business_id
 FROM business
 WHERE city="North Redington Bch"
 ORDER BY review_count desc
 limit 1)
order by is_open desc
```
