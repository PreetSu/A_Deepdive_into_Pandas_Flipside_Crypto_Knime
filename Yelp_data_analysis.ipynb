## **Database** **Modelling**

**Creating a separate entity for 'Category', 'Attribute' & 'Hours' from the 'Business' data**

```python
categories_series = business['categories']
categories = pd.DataFrame({'category': categories_series.values})
#categories['category'] = categories['category'].astype(str).str.split(',')
#categories = categories.explode('category')
categories['C_id'] = ['C{}'.format(i) for i in range(1, len(categories) + 1)]
categories['business_id'] = business['business_id']
categories.head(20)
```

```python
attributes_series = business['attributes']
attributes = pd.DataFrame({'attribute': attributes_series.values})
def extract_attributes(value):
    if isinstance(value, str):
        return {value: 'True'}
    elif isinstance(value, dict):
        return value
    else:
        return None
#attributes['attribute'] = attributes['attribute'].apply(extract_attributes)
#attributes = pd.concat([attributes.drop(['attribute'], axis=1), attributes['attribute'].apply(pd.Series)], axis=1)
attributes['A_id'] = ['A{}'.format(i) for i in range(1, len(attributes) + 1)]
attributes['business_name'] = business['name']
attributes['business_id'] = business['business_id']
attributes.head()
```

```python
hours_series = business['hours']
hours = pd.DataFrame()
hours['business_id'] = business['business_id']
days = []
for i, row in enumerate(hours_series):
    if row is not None and isinstance(row, dict):
        for day, time_range in row.items():
            days.append({'day': day, 'time_range': time_range, 'business_id': business['business_id'][i]})
hours = pd.DataFrame(days)
hours['H_id'] = ['H{}'.format(i) for i in range(1, len(hours) + 1)]
hours.head(20)
```

```python
#dropping attributes, categories, hours from business
business.drop(columns = ['hours','attributes','categories'],axis = 1,inplace = True)
```

**Creating a separate entity for 'friends', 'elite' and 'compliment' data from 'user'**

```python
friends_series = user['friends']
friends = pd.DataFrame({'friends': friends_series.values})
friends['friends'] = friends['friends'].astype(str).str.split(',')
friends = friends.explode('friends')
friends['F_id'] = ['F{}'.format(i) for i in range(1, len(friends) + 1)]
friends['user_id'] = user['user_id']

friends.head(20)
```

```python
elite_series = user['elite']
elite = pd.DataFrame({'elite': elite_series.values})
elite['elite'] = elite['elite'].astype(str).str.split(',')
elite = elite.explode('elite')
elite['E_id'] = ['E{}'.format(i) for i in range(1, len(elite) + 1)]
elite['user_id'] = user['user_id'].iloc[0]

elite.head(20)
```

```python
# Extract the compliment fields
compliments = ['compliment_hot', 'compliment_more', 'compliment_profile', 'compliment_cute', 'compliment_list',
               'compliment_note', 'compliment_plain', 'compliment_cool', 'compliment_funny', 'compliment_writer', 'compliment_photos']

# Create a DataFrame for compliments
compliment_df = user[['user_id'] + compliments]

# Generate unique compliment IDs
compliment_df['C_id'] = ['C{}'.format(i) for i in range(1, len(compliment_df) + 1)]

# Reorder columns if needed
compliment_df = compliment_df[['C_id', 'user_id'] + compliments]

compliment_df.head(20)
```

# **SQL queries to set up a new restaurant**

```python
%%sql
#1. Find the top 10 cities with the maximum number of restaurants in state. This query can help comprehend the original restaurant market in each city and state.
SELECT  state, city,count(b.business_id) AS num_bus
FROM business b join categories c on b.business_id = c.business_id
and category like '%Restaurants%'
GROUP BY  state,city
ORDER BY num_bus DESC
limit 10;
```

Philadelphia has the most restuarants. Therefore, we will focus only on that city.

```python
%%sql
#2. Find the category of restaurants with the highest stars in different cities.
select name, category,stars
from business b join categories c on b.business_id = c.business_id
where category LIKE "%Restaurants%"
and city in ("Philadelphia")
ORDER BY stars DESC
limit 15;
```

Pizzas and Delis receive high stars frequently.

```python
%%sql
#3. Finding most reviewed categories of restaurants
select sum(review_count) as total_reviews, name,category
from business b join categories c on b.business_id = c.business_id
where category LIKE "%Restaurant%"
and city in ("Philadelphia")
group by 2,3
order by 1 desc
limit 15;
```

```python
%%sql
#4. Seems like Pizza and Delis  are pretty highly rated and fall in top 10 most reviewed. Find total such restaurants.
select count(distinct category) as total_restaurants, SUM(case when category LIKE "%Delis%"  then 1 else 0 end)  as total_delis,
SUM(case when category LIKE "%Pizza%"  then 1 else 0 end)  as total_pizza
from business b join categories c on b.business_id = c.business_id
where category LIKE "%Restaurants%"
and city in ("Philadelphia")
ORDER BY 1 DESC
limit 15;
```

#### Delis are fewer compared to pizza places but are pretty highly rated, hence, we will open a deli restaurant

```python
%%sql
#5. find place in PA where people go the most (after "2022-01-01")
select count(distinct ch.business_id) as total_checkins,b.postal_code
from
business b join checkin ch
on
b.business_id = ch.business_id
join categories c on b.business_id = c.business_id
where category LIKE "%Restaurants%"
and city in ("Philadelphia")
and date(ch.date) > "2022-01-01"
group by 2
ORDER BY 1 DESC
limit 15;
```

Most checkins have happened in the area: 19107 and hence, we will go ahead and open a deli within the same area.  

```python
%%sql
#6. Finding the attributes of most tipped deli in the area
select attribute,count(distinct t.business_id	) as total_tips
from
business b join attributes a
on
b.business_id = a.business_id
join tip t on b.business_id = t.business_id
join categories c on b.business_id = c.business_id
where category LIKE "%Restaurants%"
and city in ("Philadelphia")
and  category LIKE "%Delis%"
and postal_code = 19107
group by 1
ORDER BY 1 DESC
limit 15;
```

Seems like casual delis are the most highly tipped. We want to understand if it is an acceptable business model to not allow BYOB

```python
#.7  Finding total deli where BYOB is False
%%sql
select attribute,count(distinct t.business_id	) as total_tips
from
business b join attributes a
on
b.business_id = a.business_id
join tip t on b.business_id = t.business_id
join categories c on b.business_id = c.business_id
where category LIKE "%Restaurants%"
and city in ("Philadelphia")
and  category LIKE "%Delis%"
and postal_code = 19107
and JSON_EXTRACT(attribute, "$.BYOB") = "False"
group by 1
ORDER BY 1 DESC
limit 10;
```

We can use similar attributes for our restaurant.

```python
#Query 8 Finding the most common hours of operations for  delis
%%sql
select time_range, count(*)
from business b  join categories c on b.business_id = c.business_id
join hours h on h.business_id = b.business_id
where category LIKE "%Restaurant%"
and city in ("Philadelphia")
and  category LIKE "%Delis%"
group by 1
order by 2 desc
limit 10;
```

Our restaurant can operate between the standard hours of 7 am to 10pm

```python
#Query 9 Finding reviews about deli restuarants to take recommendations from low rated restaurants
%%sql
select text from review where (text like '%deli %' or text like '%Deli %') and stars < 4
limit 5;
```

It looks like reviewers are particularly upset about slow service in delis. We need to make sure we correct that in our business
